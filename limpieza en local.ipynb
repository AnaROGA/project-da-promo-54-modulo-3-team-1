```python
roles_dict = {
    "sales executive": "sales",
    "research scientist": "research & development",
    "laboratory technician": "research & development",
    "healthcare representative": "research & development",
    "manufacturing director": "research & development",
    "sales representative": "sales",
    "research director": "research & development",
    "human resources": "human resources",
    "manager": "sales"  #dejamos solo manager - sales basándonos en que es el que más veces aparece
}
```

```python
# Normalizar columna jobrole
df_copia["jobrole_clean"] = df_copia["jobrole"].str.strip().str.lower()
# Normalizar claves del diccionario también
roles_dict_clean = {k.lower().strip(): v for k, v in roles_dict.items()}
# Aplicar el mapping
df_copia["department"] = (
    df_copia["jobrole_clean"].map(roles_dict_clean)
    .fillna(df_copia["department"])
)
```

```python
# Lista de columnas a limpiar
columnas_monetarias = ['monthlyincome', 'monthlyrate', 'salary']

# Iterar sobre las columnas y aplicar las transformaciones
for columna in columnas_monetarias:
    # 1. Convertir a string para aplicar las operaciones
    df_copia[columna] = df_copia[columna].astype(str)
    
    # 2. Eliminar el símbolo '$'
    df_copia[columna] = df_copia[columna].str.replace('$', '', regex=False)
    
    # 3. Reemplazar la coma (,) por el punto (.)
    df_copia[columna] = df_copia[columna].str.replace(',', '.', regex=False)
    
    # 4. Convertir la columna al tipo float
    # El 'errors='coerce'' convertirá valores que no sean numéricos (como 'NaN' o cadenas vacías) a NaN.
    df_copia[columna] = pd.to_numeric(df_copia[columna], errors='coerce')

print("¡Columnas monetarias limpiadas y convertidas a float!")

# Mostrar un resumen de las columnas transformadas (las primeras filas y sus tipos)
print(df_copia[columnas_monetarias].head())
print(df_copia[columnas_monetarias].info())
```

```python
# Lista de columnas constantes e inútiles para el análisis predictivo/explicativo
columnas_a_eliminar = ['employeecount', 'standardhours', 'over18']

# Eliminar las columnas
df_copia = df_copia.drop(columns=columnas_a_eliminar)

print("\n¡Columnas constantes e inútiles eliminadas!")
print(f"Las columnas eliminadas son: {columnas_a_eliminar}")
```

```python
# Mapeo de valores que representan 'Sí'
mapeo_si = [1, '1', True, 'True', 'Yes', 'yes']

# Mapeo de valores que representan 'No'
mapeo_no = [0, '0', False, 'False', 'No', 'no']

#Convertir la columna a string para asegurar que las operaciones de reemplazo funcionen para todos los tipos
df_copia['remotework'] = df_copia['remotework'].astype(str)

#Asignar 'Sí' a todos los valores positivos
df_copia['remotework'] = df_copia['remotework'].apply(
    lambda x: 'Sí' if x in [str(val).lower() for val in mapeo_si] else x.lower()
)

#Asignar 'No' a todos los valores negativos
df_copia['remotework'] = df_copia['remotework'].apply(
    lambda x: 'No' if x in [str(val).lower() for val in mapeo_no] else x.lower()
)

# Ahora, reemplazamos la cadena 'nan' (que proviene de los NaN originales convertidos a string)
df_copia['remotework'] = df_copia['remotework'].replace('nan', None) # O np.nan si importaste numpy

print("Columna 'remotework' estandarizada a 'Sí' y 'No'.")
print("\nNuevo Conteo de Valores (incluyendo posibles NaN):")
print(df_copia['remotework'].value_counts(dropna=False))
```

```python
#normalizacion de bussinestravel
# Mapeo de valores simplificados para mejor lectura
mapeo_viajes = {
    'travel_rarely': 'Rarely',
    'travel_frequently': 'Frequently',
    'non-travel': 'No Travel'
}

#Convertir la columna a minúsculas y reemplazar el guion bajo por un espacio
df_copia['businesstravel'] = df_copia['businesstravel'].astype(str).str.lower().str.replace('_', ' ', regex=False)

# Aplicar el mapeo de simplificación
df_copia['businesstravel'] = df_copia['businesstravel'].replace(mapeo_viajes, regex=True)

# Aplicar el Formato de Título
df_copia['businesstravel'] = df_copia['businesstravel'].str.title()

#Los NaN originales deben volver a ser nulos
df_copia['businesstravel'] = df_copia['businesstravel'].replace('Nan', None)

print("Columna 'businesstravel' estandarizada, simplificada y con Formato Título.")
print("\nNuevo Conteo de Valores:")
print(df_copia['businesstravel'].value_counts(dropna=False))
```

```python
# normalizacion remotework nuevamente
# Lista de todos los valores que deben ser 'Sí', en minúsculas
valores_si = ['sí', 'si', 'yes', 'true', '1']

# Lista de todos los valores que deben ser 'No', en minúsculas
valores_no = ['no', 'false', '0'] 

# Convertir a string, a minúsculas, y eliminar espacios extra 
df_copia['remotework'] = df_copia['remotework'].astype(str).str.lower().str.strip()

#Reemplazar todos los valores positivos por 'Sí'
df_copia['remotework'] = df_copia['remotework'].replace(valores_si, 'Sí')

#Reemplazar todos los valores negativos por 'No'
df_copia['remotework'] = df_copia['remotework'].replace(valores_no, 'No')

#Convertir los nulos restantes a NaN de Pandas
df_copia['remotework'] = df_copia['remotework'].replace(['nan', 'none'], np.nan)

print("Columna 'remotework' estandarizada a 'Sí' y 'No'.")
print("Los valores nulos no han sido imputados.")

# Mostrar el nuevo conteo para verificar
print("\nNuevo Conteo de Valores Final:")
print(df_copia['remotework'].value_counts(dropna=False))
```

```python
#Gestión de Variables Categóricas:Imputación por "Unknown"
#Las variables categóricas con altos porcentajes de NaN (entre el 40% y 48%) son demasiado altas para ser imputadas con la moda. 
#La mejor estrategia es reemplazarlas con la etiqueta "Unknown" (Desconocido). Esto conserva la información de que el dato faltó, 
#sin sesgar la distribución existente.

columnas_categoricas_unknown = ['businesstravel', 'educationfield', 'overtime', 'maritalstatus']

for col in columnas_categoricas_unknown:
    # Aseguramos que la columna sea tipo Object (string) para que acepte la nueva etiqueta
    df_copia[col] = df_copia[col].astype('object').fillna('Unknown')

print(f"Nulos en variables categóricas imputados con 'Unknown'.")
```

```python
#Gestión de Variables Numéricas: Imputación por Mediana
#Las columnas monetarias tienen porcentajes significativos de NaN. Para preservarlas y evitar sesgos por outliers (valores atípicos), 
#usaremos la mediana para la imputación.

columnas_numericas_mediana = ['monthlyincome', 'salary']

for col in columnas_numericas_mediana:
    mediana_valor = df_copia[col].median()
    df_copia[col] = df_copia[col].fillna(mediana_valor)

print(f"Nulos en variables monetarias imputados con la Mediana.")
```

```python
print("\nVerificación de Nulos después de la imputación:")
print(df_copia.isnull().sum().sum())
```
